Looking at FiveThirtyEight's data on which US State has the worst drivers. The data set includes summary level information about fatal accidents as well as insurance premiums for each state as of 2010.

```
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```
df = pd.read_csv('https://raw.githubusercontent.com/VictorOmondi1997/visualizing_us_worst_drivers/master/us_worst_drivers.csv')
df.head()
```

# Regression plot

# regression plot of premiums vs. insurance_losses

```
sns.regplot(data=df, x='insurance_losses', y='premiums')
plt.show()
```

```
sns.lmplot(data=df, x='insurance_losses', y='premiums')
plt.show()
```

This data set includes geographic information by state and area. It might be interesting to see if there is a difference in relationships based on the `Region` of the country.

```
sns.lmplot(data=df, x='insurance_losses', y='premiums', hue='Region')
plt.show()
```

The results are bit difficult to read. Maybe using multiple lines is not the best approach. Let's try another one.

```
sns.lmplot(data=df, x='insurance_losses', y='premiums', row='Region')
plt.show()
```
